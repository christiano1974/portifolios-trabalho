```python
### Este programa verifica se um número é palíndromo
### um número é palíndromo quando ao inverter os algorismos
### o número não se altera. 
### Por exemplo: 101, 3223, 444, assim por diante
### autor Christiano Garcia   data: 05/11/2021
```

```python
num = int(input("Digite um número inteiro: \n"))
num = str(num)
inverso = ' '
for k in range(len(num)):
    inverso += num[len(num)-1-k]
num = int(num)
inverso = int(inverso)
if num == inverso:
    print('O número digitado %d é um número palíndromo !!!\n' % num)
else:
    print('O número digitado %d não é um número palíndromo !!!\n' %num)    
```
